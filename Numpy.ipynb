#### Numpy

```python
import numpy as np
import time
```

```python
#lets declare a size
size = 100000000

#create two lists
list1 = range(size)
list2 = range(size)

#create two numpy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)
```

```python
#calculate time for python list
start = time.time()
result1 = [(x+y) for x,y in zip(list1,list2)]
print('Time for python lists in msec : ',(time.time() - start))

#calculting for numpy array
start = time.time()
result2 = (arr1 + arr2)
print('Time for numpy array in msec : ',(time.time() - start))
```

#### Q3. Hpw to replace all values greater than a given value or a cutoff

```python
array1 = np.arange(start = 12, stop=240, step = 12)
array1
```

```python
#method 1
output_array = np.clip(array1 ,a_min=50, a_max=200)
output_array
```

#### 6. Find the common and uncommon elements

```python
arr1 = np.array(range(7))
arr2 = np.array(range(5,10))
```

```python
# Print only the common numbers
# Print only the uncommon numbers
print(arr1)
print(arr2)
```

```python
#common element
np.intersect1d(arr1,arr2)
```

```python
# uncommon values
np.setdiff1d(np.union1d(arr1,arr2),np.intersect1d(arr1,arr2))
```
